```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from afinn import Afinn
```

```python
directory = '/Users/abhishekkumar/Documents/programming/himym/data/dataframes/'

main_charachters = ['ted', 'robin', 'barney', 'marshall', 'lily']
sentiments = pd.DataFrame(columns=['Season', 'Speaker', 'Message'])

for file in os.listdir(directory):
    fileName = os.fsdecode(file)
    season = pd.read_csv(directory+file)
    for row in season.itertuples():
        speaker = str(row[4]).lower()
        if speaker in main_charachters:
            sentiments = sentiments.append(
                {
                    'Season': row[1],
                    'Speaker': speaker,
                    'Message': str(row[5])[:-1]
                }, ignore_index=True)

sentiments
```

```python
set(sentiments['Season'])
```

```python
af = Afinn()
score = sentiments['Message'].map(af.score) 
sentiments['Sentiment Score'] = score
```

```python
by_season = pd.DataFrame(columns=['Season', 'Average Sentiment', 'Std Dev Sentiment', 'Sentiment Range'])
for i in range(1,10):
    cur_season = sentiments.where(sentiments['Season']==i).dropna()
    by_season = by_season.append(
        {
            'Season':i,
            'Average Sentiment': np.mean(cur_season['Sentiment Score']),
            'Std Dev Sentiment': np.std(cur_season['Sentiment Score']),
            'Sentiment Range': max(cur_season['Sentiment Score']) - min(cur_season['Sentiment Score'])
        }, ignore_index=True)

by_season
```

```python
by_charachter = pd.DataFrame(columns=['Charachter', 'Average Sentiment', 'Std Dev Sentiment', 'Sentiment Range'])
main_charachters = ['ted', 'robin', 'barney', 'marshall', 'lily']
for charachter in main_charachters:
    cur_charachter = sentiments.where(sentiments['Speaker']==charachter).dropna()
    by_charachter = by_charachter.append(
        {
            'Charachter':charachter,
            'Average Sentiment': np.mean(cur_charachter['Sentiment Score']),
            'Std Dev Sentiment': np.std(cur_charachter['Sentiment Score']),
            'Sentiment Range': max(cur_charachter['Sentiment Score']) - min(cur_charachter['Sentiment Score'])
        }, ignore_index=True)

by_charachter
```

```python
charachter_and_season = pd.DataFrame(columns=['Season', 'Charachter', 'Average Sentiment', 'Std Dev Sentiment', 'Sentiment Range'])

for i in range(1,10):
    for charachter in main_charachters:
        cur_season = sentiments.where((sentiments['Season']==i) & (sentiments['Speaker']==charachter)).dropna()
        charachter_and_season = charachter_and_season.append(
            {
                'Season':i,
                'Charachter': charachter,
                'Average Sentiment': np.mean(cur_season['Sentiment Score']),
                'Std Dev Sentiment': np.std(cur_season['Sentiment Score']),
                'Sentiment Range': max(cur_season['Sentiment Score']) - min(cur_season['Sentiment Score'])
            }, ignore_index=True)

charachter_and_season
```

```python
i = -.3
plt.figure(figsize=(20,10))
for charachter in main_charachters:
    plt.bar(np.arange(1, 10)-i, charachter_and_season.where
                                        (charachter_and_season['Charachter']==charachter)
                                        .dropna()['Average Sentiment'], width=.15)
    i+= .15
```
