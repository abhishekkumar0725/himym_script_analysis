```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
import screenTime

charachterLines = pd.DataFrame(columns=['season', 'ted', 'robin', 'barney', 'marshall', 'lily'])
directory = '/Users/abhishekkumar/Documents/programming/himym/data/dataframes/'
for file in os.listdir(directory):
    fileName = os.fsdecode(file)
    season = pd.read_csv(directory+file)
    seasonDict = screenTime.lineCounter(season)
    seasonDict['season'] = season.iloc[0]['Season']
    charachterLines = charachterLines.append(seasonDict,ignore_index=True)
        
charachterLines = charachterLines.sort_values(by=['season'])
charachterLines
```

```python
charachterLines.plot(y=['ted', 'robin', 'barney', 'marshall', 'lily'], x='season')
```

```python
charachterLines.plot(kind='bar',y=['ted', 'robin', 'barney', 'marshall', 'lily'], x='season')
```

```python
screentimePercentage = pd.DataFrame(columns=['season', 'ted', 'robin', 'barney', 'marshall', 'lily'])

for row in charachterLines.itertuples():
    lines = sum(row[1:])
    screentimePercentage = screentimePercentage.append(
        {
            'season': row[1],
            'ted': row[2]/lines,
            'robin': row[3]/lines,
            'barney': row[4]/lines,
            'marshall': row[5]/lines,
            'lily': row[6]/lines
        }, ignore_index=True)

screentimePercentage
```

```python
screentimePercentage.plot(y=['ted', 'robin', 'barney', 'marshall', 'lily'], x='season')
```

```python
screentimePercentage.plot(kind='bar',y=['ted', 'robin', 'barney', 'marshall', 'lily'], x='season')
```
